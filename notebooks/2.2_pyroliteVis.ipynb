```python
import numpy as np
import pyrolite.plot
import pyrolite.geochem
import matplotlib.pyplot as plt
from scripts.dataloader import load_belts
```

```python
df = load_belts()
df.head(2)
```

```python
df.pyrochem.REE.pyrochem.normalize_to("PM_PON").pyroplot.spider(color="k", alpha=0.02)
```

```python
df[["FeO", "MgO", "Al2O3"]].pyroplot.ternary(c="k", alpha=0.1)
```

```python
df[["FeO", "MgO", "Al2O3"]].pyroplot.density(bins=100)
```

```python
df[["FeO", "MgO", "Al2O3"]].pyroplot.heatscatter(bins=100, s=5)
```

```python
from pyrolite.util.plot import mappable_from_values

alpha = 0.5
color_by = df.Zr

ax = df[df.pyrochem.list_compositional].pyroplot.parallel(
    figsize=(14, 5), color_by=color_by, cmap="viridis", alpha=alpha
)
plt.colorbar(mappable_from_values(color_by), alpha=alpha, label=color_by.name)
ax.set_yscale('log')
```
